```python
from sklearn import preprocessing
some=preprocessing.LabelEncoder()
data['Sex']=some.fit_transform(data['Sex'])
data['Embarked']=some.fit_transform(data['Embarked'].astype(str))

```

```python
y=data.Survived.values
data=data.drop(['Survived',],axis=1)
x=data
```

```python
from sklearn.model_selection import train_test_split
x_tr,x_te,y_tr,y_te=train_test_split(x,y,train_size=0.8,random_state=3)
```

```python
x_tr=np.matrix(x_tr)
x_te=np.matrix(x_te)
```

```python
x_tr.shape,x_te.shape,y_tr.shape,y_te.shape
```

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
model = Sequential()
```

```python
y_tr=keras.utils.to_categorical(y_tr,2)
y_te=keras.utils.to_categorical(y_te,2)
```

```python
model.add(Dense(units=32,activation='relu',input_shape=(4,)))
model.add(Dropout(0.2))
model.add(Dense(units=16,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(units=14,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(units=10,activation='relu'))
model.add(Dropout(0.2))


model.add(Dense(units=8,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(units=6,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(units=4,activation='relu'))
model.add(Dropout(0.2))


model.add(Dense(units=2,activation='softmax'))
```

```python
model.compile(optimizer="Adam",loss="categorical_crossentropy",metrics=['accuracy'])
```

```python
my_model=model.fit(x_tr,y_tr,batch_size=32,epochs=100)
```

```python
model.evaluate(x_te,y_te,batch_size=64)
```

```python
ans1=model.predict(x_tr)
ans1
```
